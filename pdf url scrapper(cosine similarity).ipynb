```python
from langchain.document_loaders import PyPDFLoader,UnstructuredURLLoader
```

```python
loader=PyPDFLoader("abhijeet rajesh nikam re.pdf")
pdf_data=loader.load()
```

```python
pdf_data
```

```python
loader=UnstructuredURLLoader(urls=["https://www.britannica.com/technology/automobile"])
```

```python
data=loader.load()
```

```python
data
```

```python
combined_data=data+pdf_data
```

```python
combined_data
```

```python
from sentence_transformers import SentenceTransformer
```

```python
embed=SentenceTransformer("all-MiniLM-L6-v2")
```

```python
data[0].page_content
```

```python
(combined_data)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
spliter=RecursiveCharacterTextSplitter(separators=['\n','.',' '],chunk_size=1000,chunk_overlap=100)
chunks=[spliter.split_documents([d])for d in combined_data]
```

```python
chunks_flatten=[chunk for sublist in chunks for chunk in sublist]
```

```python
len(chunks_flatten)
```

```python
vectors=[embed.encode(chunk.page_content)for chunk in chunks_flatten]
```

```python
[print(chunk.page_content)for chunk in chunks_flatten]
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x=embed.encode('DATA ENGINEER OR DATA SCIENTIST IN TATA MOTORS')
```

```python
x_reshaped = x.reshape(1, -1)
```

```python
co=cosine_similarity(x_reshaped,vectors)
```
